```python
# 도소매조사구분명에 따라 데이터프레임 나누기
df_dm = df[df['EXMN_SE_NM'] == '도매']   # 도매
df_sm = df[df['EXMN_SE_NM'] == '소매']   # 소매
df_07 = df[df['EXMN_SE_NM'] == '07']     # 대형마트, 백화점, 전문점 등
df_08 = df[df['EXMN_SE_NM'] == '08']     # 슈퍼마켓
df_09 = df[df['EXMN_SE_NM'] == '09']     # 온라인
```



*   target_a : 조사구분명이 '도매'인 것들의 월별 팽이버섯 평균 품목가격 확인 ( 5kg 당 가격 )
*   target_b : 조사구분명이 '소매'인 것들의 월별 팽이버섯 평균 품목가격 확인 ( 150g 당 가격 )
*   target_c : 조사구분명이 '온라인'인 것들의 월별 팽이버섯 평균 품목가격 확인 ( 150g 당 가격 )





```python
# 품종명
target = '팽이버섯'
n = 33.3

# 도매, 소매, 온라인의 월별 품목 평균 가격
target_a = df_dm[(df_dm['PDLT_NM']==target)].groupby('PRCE_REG_MONTH')['PDLT_PRCE'].mean()
target_b = df_sm[(df_sm['PDLT_NM']==target)].groupby('PRCE_REG_MONTH')['PDLT_PRCE'].mean() * n   # 150(g) x 33.3 = 5(kg)
target_c = df_09[(df_09['PDLT_NM']==target)].groupby('PRCE_REG_MONTH')['PDLT_PRCE'].mean() * n   # 150(g) x 33.3 = 5(kg)

# 데이터프레임으로 합쳐서 평균 가격 비교
target_compare = pd.concat([target_a, target_b, target_c], axis=1)
target_compare.columns = ['Wholesale', 'Retail', 'Online']

print(target_compare)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 막대그래프 그리기 함수
def plot_bar_chart(target_compare):
    # 그래프 기본 설정
    target_compare.plot(kind='bar', figsize=(12, 6), width=0.8)

    # 그래프 제목 및 레이블 설정
    plt.title('Monthly Price Comparison', fontsize=16)
    plt.xlabel('MONTH', fontsize=14)
    plt.ylabel('Average Price', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)  # x축 레이블 회전
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # 그래프 출력
    plt.tight_layout()  # 레이아웃 조정
    plt.show()

# 함수 호출
plot_bar_chart(target_compare)
```

### 변동계수 구하기

```python
def get_cv(price):
    mean = price.mean()
    std = price.std()
    return std / mean * 100 if mean != 0 else 0
```

### Market별 품종 변동계수 그래프 출력 함수

```python
def plot_cv_graph(pdlt, spcs, df, threshold):

  # 명시된 품종을 판매하는 MRKT_NM 추출
  markets = df[(df['PDLT_NM'] == pdlt) & (df['SPCS_NM'] == spcs)]['MRKT_NM'].unique()

  # Market별 그래프 출력
  for market in markets:

    # 시장별 데이터 필터링
    market_data = df[(df['PDLT_NM'] == pdlt) & (df['SPCS_NM'] == spcs) & (df['MRKT_NM'] == market)]

    # 월별 변동계수 계산
    cv_data = market_data.groupby('PRCE_REG_MONTH')['PDLT_PRCE'].apply(get_cv).reset_index()
    cv_data.rename(columns={'PDLT_PRCE': 'CV'}, inplace=True)

    # 월별 CV 값 그래프 출력
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=cv_data, x='PRCE_REG_MONTH', y='CV', marker=',', color='blue')

    # 안정성 구분 표시
    plt.scatter(cv_data['PRCE_REG_MONTH'][cv_data['CV'] <= threshold],
                cv_data['CV'][cv_data['CV'] <= threshold],
                color='green', label=f'Stable (CV ≤ {threshold}%)', s=100)
    plt.scatter(cv_data['PRCE_REG_MONTH'][cv_data['CV'] > threshold],
                cv_data['CV'][cv_data['CV'] > threshold],
                color='red', label=f'Unstable (CV > {threshold}%)', s=100)

    # 그래프 제목 및 레이블 설정
    plt.xlabel('Month', fontsize=14)
    plt.ylabel('CV (%)', fontsize=14)
    plt.axhline(y=threshold, color='blue', linestyle='--', label=f'{threshold}% Threshold')
    plt.title(f'{pdlt} ({spcs}) CV in {market}', fontsize=16) # MRKT_NM 추가
    plt.legend(fontsize=12)
    plt.grid(True)

    # 그래프 출력
    plt.show()
```

### 도매점별 샤인머스켓 가격 변동성 확인

```python
plot_cv_graph('사과', '후지', df[df['EXMN_SE_NM'] == '도매'], 15)
```
